```python
data_path = 'PBMC_train/data10k/'
out_path = data_path
adata = sc.read(data_path + 'matrix.mtx').T
adata.var_names = pd.read_csv(data_path + 'features.tsv', header=None, sep='\t')[1]
adata.obs_names = pd.read_csv(data_path + 'barcodes.tsv', header=None)[0]
adata.var_names_make_unique()
sc.pp.filter_cells(adata, min_genes=500)
sc.pp.filter_genes(adata, min_cells=5)
adata.obs['n_counts'] = adata.X.sum(axis=1)
sc.pl.scatter(adata, x='n_counts', y='n_genes')
adata = adata[adata.obs['n_genes'] < 3000, :]
adata = adata[adata.obs['n_counts'] < 13000, :]
adata.raw = sc.pp.log1p(adata, copy=True)
df = pd.DataFrame(adata.X.todense())
df.columns = adata.var.index
df.to_csv(out_path + "data10k_counts_all.txt", sep="\t")
sc.pp.normalize_per_cell(adata)
df = pd.DataFrame(adata.X.todense())
df.columns = adata.var.index
df.to_csv(out_path + "data10k_norm_counts_all.txt", sep="\t")
filter_result = sc.pp.filter_genes_dispersion(adata.X, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.filter_genes_dispersion(filter_result)
adata = adata[:, filter_result.gene_subset]
sc.pp.normalize_per_cell(adata)
adata.X_noscale = adata.X
sc.pp.log1p(adata)
sc.pp.regress_out(adata, ['n_counts'])
sc.pp.scale(adata)
sc.pp.neighbors(adata)
```

```python
sc.tl.umap(adata)
sc.tl.louvain(adata, resolution=1)
sc.pl.umap(adata, color='louvain')
```

```python
sc.pl.stacked_violin(adata,['MS4A1','CD8A','CD8B','CD4','IL7R', 'CLEC4C',
                            'GNLY','NKG7', 'FCER1A','CST3','CD14', 'LYZ', 'FCGR3A'],groupby='louvain')
```

```python
new_celltypes = ['CD4Tcells', 'CD4Tcells2', 'Monocytes', 'Bcells', 'Monocytes2', 'NK', 'NK2',
                 'NK3', 'Bcells2', 'CD8Tcells', 'Monocytes3', 'Unknown', 'CD4Tcells3', 'Dendritics']
adata.rename_categories('louvain', new_celltypes)
sc.pl.umap(adata, color='louvain', legend_loc='on data')
adata.rename_categories('louvain', new_celltypes)
sc.pl.umap(adata, color='louvain', legend_loc='on data')
# Make new category
celltypes = pd.DataFrame(adata.obs['louvain'])
celltypes.louvain.replace(['CD4Tcells2', 'Monocytes2', 'NK2', 'NK3', 'CD4Tcells3', 'CD4Tcells4', 'Bcells2', 'Monocytes3'],
                          ['CD4Tcells', 'Monocytes', 'NK', 'NK', 'CD4Tcells','CD4Tcells', 'Bcells', 'Monocytes'], inplace=True)
adata.obs['celltype'] = celltypes.louvain
sc.pl.umap(adata, color='celltype', legend_loc='on data')
celltypes = pd.DataFrame(adata.obs['celltype'])
celltypes.columns = ['Celltype']
celltypes.to_csv(out_path + "data10k_celltypes.txt", sep="\t")
```
